```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df11=pd.read_csv('DP_LIVE_01042020222431768.csv')
df12=df11.loc[:,['LOCATION','Value']].groupby(['LOCATION']).sum(axis=0)

plt.figure(figsize=(15,5))

df12.plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Number of Road Accidents ")
plt.title("Number of Road Accidents  Country")


plt.xticks(rotation=90, fontsize=7)
plt.legend()


plt.tight_layout()
plt.show()
```
